**MATPLOTLIB LIBRARY**

```python
import numpy as np
import matplotlib.pyplot as plt 
```

***LINEAR GRAPH***

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y)
#to only see the graph
plt.show()
```

```python
#to change color of line
x=[1,2,3,4]
y=[5,6,7,8]
c='r'#here r is red color
plt.plot(x,y,c)
#to only see the graph
plt.show()
```

```python
x=[5,2,9,4,7]
y=[3,6,8,5,4]
c='g'
plt.plot(x,y,c)
#show function removes "[<matplotlib.lines.Line2D at 0x220b2755ed0>]"
plt.show()
```

```python
#To point the markers/points in linear graph
x=[5,2,9,4,7]
y=[3,6,8,5,4]
c='g'
plt.plot(x,y,c,marker='o')
plt.show()
```

```python
#to get the graph along x and y axis seperately as well y should have markers.
x=[5,2,9,4,7]
y=[3,6,8,5,4]
c='g'
plt.plot(x,y,c)
#to draw graph along y-axis using marker assigned
plt.plot(y,marker='>',c='r')
#to draw graph along x-axis using marker assigned
plt.plot(x,marker='o',c='y')
plt.show()
```

***BAR GRAPH***

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
#TO CHANGE COLOR OF EVERY BARS
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
#for horizontal representation of bar graph-->barh() method is used
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
#for different in every bar
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
c=['r','g','y','b']
plt.barh(x,y,color=c)
plt.show()
```

```python
#for same color in every bar
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
c='r'
plt.barh(x,y,color=c)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color="hotpink")
plt.show()
```

```python
#to change the size of width in bar graph
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,width=0.5)
plt.show()
```

```python
#to set the height of bar graph according to our requirement
#height can only be set in horizontal bar graph 
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,height=0.1)
plt.show()
```

***SCATTER PLOT***

```python
x=[2,5,3,10,4]
y=[5,9,7,8,7]
plt.scatter(x,y)
plt.show()
```

```python
#to give title name to the graph
x=[2,5,3,10,4]
y=[5,9,7,8,7]
plt.title("SCATTER PLOT")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y)
plt.show()
```

```python
#to change color 
x=[2,5,3,10,4]
y=[5,9,7,8,7]
plt.title("SCATTER PLOT")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y,color="green")
plt.show()
```

```python
#to change color for every point
x=[2,5,3,10,4]
y=[5,9,7,8,7]
c=['r','y','b','magenta','g']
plt.title("SCATTER PLOT")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y,color=c)
plt.show()
```

```python
#to change size of representation of dot
x=[2,5,3,10,4]
y=[5,9,7,8,7]
plt.title("SCATTER PLOT")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y,s=150)
plt.show()
```

***PIE CHART***

```python
x=[10,20,30,40]
y=['english','math','hindi','science']
c=["magenta","brown","aqua","yellow"]
plt.pie(x,labels=y,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
y=['english','math','hindi','science']
plt.pie(x,labels=y)
plt.show()
```
